```python
df=dframe
```

```python
df['classification'] = df['classification'].apply(lambda x:'ckd' if x=='ckd\t' else x)

df['cad'] = df['cad'].apply(lambda x:'no' if x=='\tno' else x)

df['dm'] = df['dm'].apply(lambda x:'yes' if x=='\tyes' else x)
df['dm'] = df['dm'].apply(lambda x:'yes' if x==' yes' else x)
df['dm'] = df['dm'].apply(lambda x:'no' if x=='\tno' else x)

df['rc'] = df['rc'].apply(lambda x:'5.2' if x=='\t?' else x)

df['wc'] = df['wc'].apply(lambda x:'9800' if x=='\t?' else x)
df['wc'] = df['wc'].apply(lambda x:'9800' if x=='\t8400' else x)
df['wc'] = df['wc'].apply(lambda x:'9800' if x=='\t6200' else x)

df['pcv'] = df['pcv'].apply(lambda x:'41' if x=='\t?' else x)
df['pcv'] = df['pcv'].apply(lambda x:'41' if x=='\t43' else x)
```

```python
for i in df.columns:
    print(i)
    print(set(df[i].tolist()))
    print('\n')
```

```python
df.dtypes
```

```python
df['id'] = df['id'].apply(lambda x: float(x))
df['pcv'] = df['pcv'].apply(lambda x: float(x))
df['wc'] = df['wc'].apply(lambda x: float(x))
df['rc'] = df['rc'].apply(lambda x: float(x))
```

```python
for i in df.columns:
    print(i)
    print(set(df[i].tolist()))
    print('\n')
```

```python
df.dtypes
```

```python
df_nominal = df[['id','rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane','classification']]
df_numerical = df[['id','age','bp','sg','al','su','bgr','bu','sc','sod','pot','hemo','pcv','wc','rc']]
```

```python
# Imputing null values
from sklearn.impute import SimpleImputer
imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')

# null values are replaced by mode,mean value of the column.. median can also be used
df_imputed_nominal = pd.DataFrame(imp_mode.fit_transform(df_nominal))
df_imputed_numerical = pd.DataFrame(imp_mean.fit_transform(df_numerical))

df_imputed_nominal.columns = df_nominal.columns
df_imputed_numerical.columns = df_numerical.columns

df_imputed = pd.merge(df_imputed_numerical,df_imputed_nominal, how='inner', on = 'id')
```
